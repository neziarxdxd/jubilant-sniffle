## Import 
import all the requirements such as the modules

```python
#importing libaries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

## Find HTML Tags
- Find the URL
- Create a request
- scrape using Beautiful Soup
- scrape the table body including its class 

```python
#sample url
url = "https://github.com/aakashpahwa/diabetes-dataset/blob/master/dataset_37_diabetes.csv"
#request the page
page = requests.get(url)
#creating a soup object and the parser
soup = BeautifulSoup(page.text, 'lxml')

#creating a table body to pass on the soup to find the table
table_body = soup.find('table',{"class":"js-csv-data csv-data js-file-line-container"})
table_body
```

```python
def transform(data):
  if data=="tested_positive":
    return 1
  else:
    return 0
```

# Append to List 

- find all the <TR> or Table rows in the second index [1] upto the last 
- inside the TR find all the TD 
- in the TD strip all  the white space 
- before appending make sure to get the proper data type 
- then append it 

```python
#creating an empty list to store information
row_data = []

#creating a table 
for row in table_body.find_all('tr')[1:]:
  col = row.find_all('td')
  col = [ele.text.strip() for ele in col ] # stripping the whitespaces
  row_data.append([float(col[1]),float(col[2]),float(col[3]),float(col[4]),float(col[5]),float(col[6]),float(col[7]),float(col[8]), transform(col[9])])
  #append the column
row_data
```

# DataFrame
- add row data to the dataframe 

```python
df = pd.DataFrame(row_data)
df
```

add the rowdata and the headers in the DataFrame

```python
headers = []
for i in soup.find_all('th'):
  col_name = i.text.strip().lower().replace(" ", "_")
  headers.append(col_name)
headers
```

## CONVERT DATAFRAME TO CSV
- convert to csv by using .to_csv

```python

df = pd.DataFrame(row_data, columns=headers)
df
#into a file 
df.to_csv('diabetes.csv')
```
